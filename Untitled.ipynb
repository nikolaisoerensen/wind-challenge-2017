```python
import generator as gn
import numpy as np
import matplotlib as mpl
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
axial_gen = gn.Generator()
radial_gen = gn.Generator(design='radial')

N_array 	= np.linspace(1,100,100)
#N_array 	= np.array([100])
d_array		= np.linspace(0.5,1.5,11)*10**-3#[mm]
#d_array 	= np.array([0.001])
#RL_array 	= np.linspace(1,50,10)


eta_rad,eta_ax,R_L_ax = (np.zeros([N_array.size,d_array.size]) for i in range(3))

for j,d in enumerate(d_array):
	axial_gen.coil.set_d_wire(d)
	radial_gen.coil.set_d_wire(d)
	for i,N in enumerate(N_array):
		
		axial_gen.coil.set_windings(N)
		axial_gen.compute()
		eta_ax[i][j] = axial_gen.eff
		R_L_ax[i][j] = axial_gen.R_L


		#radial_gen.coil.set_windings(N)
		#radial_gen.compute()
		#eta_rad[i][j] = radial_gen.eff
```

```python
fig, (ax1) = plt.subplots(1)
ax1.plot(N_array,eta_ax)#,N_array,eta_rad.T)
ax1.set_xlabel('N [1]')
ax1.set_ylabel('eta Generator [1]')
ax1.legend([str(d) for d in d_array*10**3])
ax1.set_ylim([0,1])
ax1.grid()

N = 64.0
i_d = 11-1
eta_N = np.interp(N,N_array,eta_ax[:,i_d])
ax1.scatter(N, eta_N)
R_L = np.round(np.interp(N,N_array,R_L_ax[:,i_d]),2)
txt = 'R_L = '+str(R_L)+' Ohm'
ax1.annotate(txt, xy=(N,eta_N))
```
